```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.figure import Figure
```

```python
beton = pd.read_excel('beton_dane.xls')
```

```python
beton.to_html('beton.html')
```

```python
# Zadanie 1
```

```python
# 1. W danym zakresie nie istnieją żadne puste (NaN) wartości
beton.isnull().sum()
```

```python
# 2. Wartości zerowe zawierają się w trzech kolumnach, zuzel wielkopiecowy,
# popiół lotny oraz upłynniacz.
beton.eq(0).sum()
```

```python
# 3. Wszystkie kolumny oprócz wieku są typu zmiennoprzecinkowego, float.
# Wiek betonu jest typu integer.
beton.info()
```

```python
# 4. Najcięższy beton: 2551.0 kg / Najlżejszy beton: 2194.6 kg
minimum = beton['suma_waga'].min()
maximum = beton['suma_waga'].max()
print(str(minimum) + ' : ' + str(maximum))
```

```python
# 5. Funkcja nunique() zwróci informację o unikalnych wartościach dla każdej kolumny
# Najwięcej unikalnych posiada kolumna 'wytrzymalosc' = 845, najmniej 'wiek_dni' = 14.
beton.nunique()
```

```python
# 6. Dzien w ktorym najczesciej byly robione badania dla betonu,
# to 28 dzien od jego powstania.
beton['wiek_dni'].mode()
```

```python
# Wykres rozrzutu zawartości cementu do wytrzymałości z podziałem na dni dokonania pomiaru
plt.figure(figsize=(10,10))
ax = sns.scatterplot(data = beton, x = beton['cement'], y = beton['wytrzymalosc'], hue = beton['wiek_dni'])

plt.show()
```
